BASIC IMAGE SEGMENTATION FOR ULTRASONOGRAPHY

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Memuat gambar
image_path ='/content/desktop-wallpaper-ultrasound.jpg'
image = cv2.imread(image_path)

# Mengubah gambar menjadi array 2D
pixel_values = image.reshape((-1, 3))

# Konversi ke tipe data float32
pixel_values = np.float32(pixel_values)

# Menentukan kriteria penghentian
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)

# Jumlah cluster (jumlah segmen yang diinginkan)
k = 3

# Melakukan k-means clustering
_, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Konversi titik tengah ke integer
centers = np.uint8(centers)

# Melakukan segmentasi gambar
segmented_image = centers[labels.flatten()]

# Mengembalikan bentuk gambar ke bentuk awal
segmented_image = segmented_image.reshape(image.shape)

# Menampilkan gambar asli dan hasil segmentasi
cv2_imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
```
